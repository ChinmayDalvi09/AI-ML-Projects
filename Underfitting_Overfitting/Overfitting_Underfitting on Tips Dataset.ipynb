```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
X = df[['total_bill']].values   # independent variable
y = df['tip'].values           # dependent variable
```

```python
model = LinearRegression()
```

```python
# Train model on data
model.fit(X, y)
```

```python
# Predict values
y_pred = model.predict(X)
```

```python
plt.scatter(X, y)          # actual data points
plt.plot(X, y_pred)        # predicted straight line
plt.title("Underfitting (Linear Model)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Use very high degree → model becomes too complex
poly = PolynomialFeatures(degree=8)
X_poly = poly.fit_transform(X)
```

```python
# Train model
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Predict values
y_pred = model.predict(X_poly)
```

```python
# Plot
# Too many curves  → Overfitting
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.title("Overfitting ")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sorted_idx = X[:,0].argsort()

# Plot
plt.scatter(X, y)
plt.plot(X[sorted_idx], y_pred[sorted_idx])
plt.title("Overfitting (High Degree)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
